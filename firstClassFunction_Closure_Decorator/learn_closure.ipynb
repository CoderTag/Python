## What is a Closure?
In Python, a closure is a function object that remembers values in enclosing scopes even if they are not present in memory.

To form a closure, you need three core conditions:

- A nested function (a function inside a function).
- The nested function must refer to a value defined in the enclosing (outer) function.
- The enclosing function must return the nested function.

Essentially, a closure allows a nested function to "carry around" an environment of variables from its parent function, even after the parent function has finished executing.

#### So what are closures good for?
- Closures can avoid the use of global values and provides some form of data hiding.
- It can also provide an object oriented solution to the problem.
- When there are few methods (one method in most cases) to be implemented in a class, closures can provide an alternate and more elegant solutions. 
- But when the number of attributes and methods get larger,better implement a class.

```python
def logger(func):
    def log_func(*args):
        logging.info(f'function {func.__name__} is called with arguments {args}')
        print(func(*args))
    return log_func

def add(*args):
    sum = 0
    for arg in args:
        sum += arg
    return sum

def multiply(*args):
    prod = 1
    for arg in args:
        prod *= arg
    return prod
```

### The above Code Breakdown: How Your Example Works
1. logger(func) is the outer function. It takes another function (func) as an argument.
2. log_func(*args) is the inner nested function. It accesses the func variable from the outer function's scope and captures any arguments (*args) passed to it.
3. The Closure: When logger returns log_func, log_func retains access to func, keeping it alive in memory.

#### The Assignment Phase
When you execute add = logger(add), you pass the original add function into logger.
- logger binds func to your original addition logic.
- It returns the log_func inner function, which is now saved to the variable add.
- The original add function is now wrapped inside a closure.

#### The Execution Phase
When you call add(1, 2, 3, 4, 5), you are actually calling the inner log_func with the tuple of arguments (1, 2, 3, 4, 5).
Because of the closure:
- log_func still knows exactly what func is (the original add logic).
- It logs the message: "function add is called with arguments (1, 2, 3, 4, 5)" to example.log.
- It calls func(*args) (the math), calculates 15, and prints it to the console.

```python
add = logger(add)
prod = logger(multiply)

add(1, 2, 3, 4, 5)
prod(1, 2, 3, 4)
```
